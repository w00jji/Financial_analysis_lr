```python
import pandas as pd

date = ["6/1", "6/2", "6/3", "6/4", "6/5"]
high = pd.Series([42800, 42700, 42050, 42950, 43000], index = date)
low = pd.Series([42150, 42150, 41300, 42150, 42350], index = date)

# 변동폭
diff = high -low
diff
```

```python
diff.max() # 최대값
```

```python
diff.idxmax() # 가장 큰 값이 있는 index
```

```python
diff.idxmin() # 가장 작은 값이 있는 index
```

```python
# 수익률
profit = high / low
profit
```

```python
# 누적수익률
profit.cumprod().iloc[-1]
```

```python
# 순위 처리
df= pd.read_excel('성적 처리.xlsx',sheet_name='Sheet1')
df['총합'] = df.iloc[:,2:].sum(axis =1 )
df.sort_values(['반','총합'],ascending = [True,False])
```

```python
# 순위(average) : 3등이 2명이면 둘다 3.5등으로 처리하고 다음 등수를 5등
df['순위']=df['총합'].rank(ascending = False, method = 'average')
df
```

```python
# 순위(average) : 3등이 2명이면 둘다 4등으로 처리하고 다음 등수를 5등
df['순위']=df['총합'].rank(ascending = False, method = 'max')
df
```

```python
# 순위(average) : 3등이 2명이면둘다 3등처리
df['순위']=df['총합'].rank(ascending = False, method = 'min')
df
```

```python
# 3등이 2명이면, 처음 발견한 3등은 3등, 다음 발견한 3등은 4등 (선착순)
df['순위']=df['총합'].rank(ascending = False, method = 'first')
df
```

```python
# min이랑 같음. 3등이 2명이면 둘다 3등처리하고 다음 등수를 4등

df['순위']=df['총합'].rank(ascending = False, method = 'dense')
df
```
